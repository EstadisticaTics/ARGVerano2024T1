un negocio de comida rápida se presenta a continuación:

```python
import numpy as np
import matplotlib.pyplot as plt

datos = [250, 790, 470, 510, 500, 330, 300, 750, 580, 740, 420, 260, 440, 770, 400, 540, 530, 330, 390, 380, 340, 460, 530, 260, 410, 360, 270, 510, 630, 330]

datos.sort()

valores_unicos, conteos = np.unique(datos, return_counts=True)
fa = conteos

fr = conteos / len(datos)
frecuencia_acumulada = np.cumsum(conteos)


plt.figure(figsize=(10, 6))
plt.hist(datos, bins=6, density=True)
plt.xlabel("Marcas de clase")
plt.ylabel("Frecuencia relativa (fr)")
plt.title("Histograma de los datos")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos 
data = [250, 790, 470, 510, 500, 330, 300, 750, 580, 740, 420, 260, 440, 770, 400, 540, 530, 330, 390, 380, 340, 460, 530, 260, 410, 360, 270, 510, 630, 330]


# Calcular la frecuencia acumulada
sorted_data = sorted(data)
cumulative_freq = np.cumsum(np.ones_like(sorted_data)) / len(sorted_data) * 100

# Crear el diagrama de ojiva
plt.figure(figsize=(10, 6))
plt.plot(sorted_data, cumulative_freq, marker='o')

plt.hist(datos, bins=6, density=True)
plt.xlabel('Marcas de Clase')
plt.ylabel('Frecuencia Acumulada')
plt.title('Diagrama de Ojiva')
plt.grid(True)

plt.show()
```
