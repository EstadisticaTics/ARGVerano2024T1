```python

#------------------------------------------ >> Primera funcion << ----------------------------------------------------

def frecuencia_absoluta(datos_entrada_arreglo):  
    lista_clases  = []                           
    lista_frec_absoluta = []                     
    for elementos in datos_entrada_arreglo:      
            if elementos not in lista_clases:    
                lista_clases.append(elementos)   
                lista_frec_absoluta.append(1)    
    else:                                        
            idx = lista_clases.index(elementos)  
            lista_frec_absoluta[idx] += 1        
    return lista_clases, lista_frec_absoluta      

#-------------------------------------------- >> Segunda funcion << -------------------------------------------------


def ordenar_arreglo(arreglo):                 
    tamaño_len = len(arreglo)                 
    
                                                           
    for elemento in range(tamaño_len):                    
        
        elemento_min_idx = elemento                        
        
        for j in range(elemento + 1, tamaño_len):          
                                                         
                                                          
            
                                                           
            if arreglo[j] < arreglo[elemento_min_idx]:             
                min_idx = j                               
                                                                     
        arreglo[elemento], arreglo[elemento_min_idx] = arreglo[elemento_min_idx], arreglo[elemento]  
                                       

    return arreglo 


#------------------------------------------ >> Tercera funcion << ----------------------------------------------------

def ordenar_lista_Fabsolutas(list_clases_originales, clases_ordenadas, fa_absolutas):
    
    list_fa_ordenada = []
    
    
    for clase in clases_ordenadas:
        
        idx = list_clases_originales.index(clase)
        
        fa = fa_absolutas[idx]
        
       
        list_fa_ordenada.append(fa)
    
   
    return list_fa_ordenada

#------------------------------------------ >> Cuarta funcion << ----------------------------------------------------
def frecuencia_relativa(frecuencia_absoluta):
    tot_datos = sum(frecuencia_absoluta) 
    frec_relativa = []  

    for frecuencia in frecuencia_absoluta: 
        frec_relativa.append((frecuencia / tot_datos) * 100) 

    return frec_relativa  


#------------------------------------------ >> Quinta funcion << ----------------------------------------------------

def frecuencia_acumulada(frecuencias_relativas):
    frec_acumulada = [] 
    contador = 0  

    for frecuencia in frecuencias_relativas:  
        contador += frecuencia  
        frec_acumulada.append(contador) 


    return frec_acumulada  

#---------------------------------------------------------------Procedimiento---------------------------------------------------------------


# Datos de entrada
datos_entrada = [2.5, 4.5, 5.8, 2.2, 5.7, 5.3, 4.0, 0.9, 2.5, 2.3, 3.2, 4.9, 2.7, 3.7, 3.8, 4.9, 2.8, 3.6, 4.2, 3.1, 7.2, 1.9, 5.9, 3.3, 2.9, 1.1, 1.7, 2.6, 3.1, 4.4]
clases, fa_abs = frecuencia_absoluta(datos_entrada)
clases_orden = ordenar_arreglo(clases)
fa_sorted = ordenar_lista_Fabsolutas(clases, clases_orden, fa_abs)
frecuencia_rel = frecuencia_relativa(fa_sorted)
frecuencia_rel_acum = frecuencia_acumulada(frecuencia_rel)


#---------------------------------------------------------------IMPRESION DE 4 COLUMNAS---------------------------------------------------------------


def imprimir_tabla(clases_orden, fa_sorted, frecuencia_rel, frecuencia_rel_acum):
   
    print('MarcaClases'.ljust(15), 'Fa'.ljust(15), 'Fr'.ljust(15), 'F acum'.ljust(15))
    print("---------------------------------------------------------------")

    for i in range(len(clases_orden)):
        print(f'Clase {clases_orden[i]}'.ljust(15), 
              str(fa_sorted[i]).ljust(15), 
              str(frecuencia_rel[i]).ljust(15), 
              str(frecuencia_rel_acum[i]).ljust(15))


imprimir_tabla(clases_orden, fa_sorted, frecuencia_rel, frecuencia_rel_acum)


#---------------------------------------------------------------IMPRESION DE TABLA COMPLETA---------------------------------------------------------------

def imprimir_tabla(clases_orden, fa_sorted, frecuencia_rel, frecuencia_rel_acum, lower_limits, upper_limits):
   
    print('Clases'.ljust(10), 'Fa'.ljust(8), 'Fr (%)'.ljust(8), 'F acumulada (%)'.ljust(14), 'Lim inf'.ljust(10), 'Lim sup'.ljust(10))
    print('------'.ljust(10), '---'.ljust(8), '---'.ljust(8), '-------------'.ljust(14), '-------'.ljust(10), '-------'.ljust(10))

    # Iterar sobre cada clase y sus datos correspondientes
    for i in range(len(clases_orden)):
        print(f' {clases_orden[i]}'.ljust(10), 
              str(fa_sorted[i]).ljust(8), 
              f'{frecuencia_rel[i]:.2f}'.ljust(8), 
              f'{frecuencia_rel_acum[i]:.2f}'.ljust(14), 
              str(lower_limits[i]).ljust(10), 
              str(upper_limits[i]).ljust(10))


datos_entrada =  [2.5, 4.5, 5.8, 2.2, 5.7, 5.3, 4.0, 0.9, 2.5, 2.3, 3.2, 4.9, 2.7, 3.7, 3.8, 4.9, 2.8, 3.6, 4.2, 3.1, 7.2, 1.9, 5.9, 3.3, 2.9, 1.1, 1.7, 2.6, 3.1, 4.4]


clases, fa_abs = frecuencia_absoluta(datos_entrada)


clases_orden = ordenar_arreglo(clases)


fa_sorted = ordenar_lista_Fabsolutas(clases, clases_orden, fa_abs)


frecuencia_rel = frecuencia_relativa(fa_sorted)
frecuencia_rel_acum = frecuencia_acumulada(frecuencia_rel)

lower_limits = [min(clases_orden)] + clases_orden[:-1]  
upper_limits = clases_orden[1:] + [max(clases_orden)]   


```
